```python
import sys
import pandas as pd
import numpy as np
import random
import matplotlib.pyplot as plt
from matplotlib import *
import warnings
import scipy
from scipy.signal import find_peaks
from scipy.misc import electrocardiogram
#from scipy.interpolate import interp1d
from datetime import datetime
from datetime import timedelta
import math
warnings.filterwarnings('ignore')
```

```python
datafile='2019_09_30__12_35_16_302764.csv'
##what is the minimum height of a 'peak'
PeakHeight=200

##how far do two peaks need to be to be called distinct peaks
distancebtwpeaks=10

##Open the data file
df=pd.read_csv(datafile,sep=',', delimiter=None, header='infer', engine='python')

##
df['datetime']= pd.to_datetime(df['python_datetime'])

##create empty numeric 'peak' column 
df['peak']=np.nan
df['peak']= pd.to_numeric(df['peak'])

for a in ['A0','A1','A2','A3','A4','A5']:
    #P=[]
    ##find peaks -- set peak height and distance between peaks
    P, _ = find_peaks(df[a], height=PeakHeight, distance=distancebtwpeaks)
    for b in P:
        #print(b)
        name=a+'peak'
        df.at[b,name]=1000
```

```python
values=['A0peak','A1peak','A2peak','A3peak','A4peak','A5peak']
T = dict()
V = dict()
AT= dict()
for item in values:
    T[item] = []
    V[item] = []
    AT[item] = []
    
    count=0
    time=0
    AbsoluteTime=0
    for index, row in df.iterrows():

        if count>0:
            if math.isnan(row[item]):
                time=time+row['ardno_usec_snce_lst_read']
                AbsoluteTime=AbsoluteTime+row['ardno_usec_snce_lst_read']
                pass
            else:
                Tsecond = time/1000000
                F=1/Tsecond
                ATsecond = AbsoluteTime/1000000
                T[item].append(F)
                V[item].append(row[item])
                AT[item].append(ATsecond)
                #print(item,row[item],time)
                AbsoluteTime=AbsoluteTime+row['ardno_usec_snce_lst_read']
                time=0
        
        ##ignore the first count
        if count==0:
            if math.isnan(row[item]):
                pass
            else:
                count=count+1    
                time=row['ardno_usec_snce_lst_read']
                AbsoluteTime=0
```

```python
import numpy as np
from numpy import convolve
import matplotlib.pyplot as plt
##specify window size
W = 10
shape = (1, W-1)
zeros_array = np.empty(shape)
zeros_array = zeros_array.flatten('F')
zeros_array[:] = numpy.nan

def movingaverage (values, window):
    weights = np.repeat(1.0, window)/window
    sma = np.convolve(values, weights, 'valid')
    return sma

peaks = ['A0peak','A1peak','A2peak','A3peak','A4peak','A5peak']
MA=dict()
for a in peaks:
    MA[a] = []
    MA[a] = movingaverage(T[a],W)
    MA[a] = np.concatenate((zeros_array, MA[a]), axis=0)
```

```python
fig, ax = plt.subplots(6,2, figsize = [25,20])
peaks = ['A0peak','A1peak','A2peak','A3peak','A4peak','A5peak']


for a in [0,1,2,3,4,5]:
    X=MA[peaks[a]].tolist()
    ax[a,1].plot(AT[peaks[a]],T[peaks[a]],marker='o', markerfacecolor='blue', markersize=5, color='skyblue',linestyle="-")
    ax[a,1].plot(AT[peaks[a]],X,marker='o', markerfacecolor='red', markersize=5, color='red',linestyle="-")
    #ax[a,1].set_ylabel('Frequency (1/s)')
    #ax[a,1].set_xlabel('Time (seconds)')   
    ax[a,1].set_ylabel('Frequency (1/s)')
for a in [0,1,2,3,4,5]:
    value=(peaks[a])
    y = value.replace('peak','')
    ax[a,0].plot(df['datetime'],df[y],marker='o', markerfacecolor='blue', markersize=5, color='skyblue',linestyle="-")
    ax[a,0].plot(df['datetime'],df[peaks[a]],marker='X', markerfacecolor='red', markersize=5, color='red')
    ax[a,0].set_ylabel('Signal')

    
ax[a,1].set_xlabel('Time (seconds)')  
ax[a,0].set_xlabel('Time')
```
